A time series can be decomposed into trend, seasonality and remainder (containing anything else in the time series). Altough a time series may contain cyclical component, we usually combine the trend and cycle into a single component called trend-cycle component (frequently called the trend).

a time series can be modeled as an additive or multiplicative decomposition. Which decomposition to use depends on the nature of the trend and seasonality:

In additive decomposition, each obsevation in a time series is given by:

>Value = Base Level + Trend + Seasonality + Error

In multiplicative decomposition, each obsevation in a time series is given by:

>Value = Base Level x Trend x Seasonality x Error

**Additive Decomposition:** is the most suitable decomposition method if the magnitude of the seasonal fluctuations, or the variation around the trend-cycle, does not vary with the level of the time series.

**Multiplicative decomposition:** is the most suitable method when the variation in the seasonal pattern, or the variation around the trend-cycle, appears to be proportional to the level of the time series.
